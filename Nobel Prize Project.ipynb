Visual gender modeling and predicting.

Analysis:
We can see here a nice growth in Nobel prize winnings by females as time goes on. If this trend continues we may see women break 10% of the wins by 2020.

```python
#Display the female proportion over time
nobel['decade'] = (np.floor(nobel['Year'] / 10) * 10).astype(int)
nobel['female_winner'] = nobel['Sex'] =="Female"
fe_decade_prop = nobel.groupby(['decade'], as_index=False)['female_winner'].mean()
sns.lmplot(x='decade',y='female_winner',data=fe_decade_prop, aspect=2, order=2)
plt.ylabel('Female Winners Proportion')
```

Visual nationality modeling and predicting of the top 10 nationalities of winners.

Analysis:
We can right away see that USA, Japan and Canada are the only ones showing increases over time in proportion of wins. But take a close look at the Y axis values, they change per country to have a "good fit" in the available space. Notice that the USA is about 12 times greater towards the later years than Cananda. Canada may be increasing but at a much slower rate than that of Japan and the USA. The other 7 countries seem to be leveling out or decreasing in proportion over time. The projections for 2017-2020 look to be: USA = 40%, Japan = 10%, UK = 8%, Canada = 3%, with the others reading simiarly from their respective graphs.

```python
# Create boolean columns for each of the top 10 winning nationalities then group and plot their 
# proportion over time using Lowess regresion line
sns.set(style='darkgrid', palette='husl')
def nation_prediction(df, countries, x_var_time):
    for i,c in enumerate(countries):
        column = c+'_born'
        df[column] = df['Birth Country'] == c
        c_prop_win = df.groupby(x_var_time, as_index=False)[column].mean()
        plt.figure(figsize=[11.5, 5.5])
        sns.regplot(x=x_var_time, y=column, data=c_prop_win, lowess=True)
nation_prediction(nobel, top_10_over, 'decade')
```
